```python
import matplotlib.pyplot as plt
import re

# Function to read the file and extract block size and time
def extract_data(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            # Regular expression to match the pattern "Block size: [number] Time: [number] ms"
            match = re.search(r'Block size: (\d+) Time: ([\d.]+) ms', line)
            if match:
                # Extract block size and time, convert them to appropriate types
                block_size = int(match.group(1))
                time = float(match.group(2))
                data.append((block_size, time))
    return data

# Example usage
filename = 'cache_blocking_times_no_opt_n10.txt'  # Replace with your actual file name
extracted_data = extract_data(filename)
print(extracted_data)


# Make a plot from extracted_data
x, y = zip(*extracted_data)
plt.plot(x, y)
plt.xlabel('Block size')
plt.ylabel('Time (ms)')
plt.title('Cache blocking times')
plt.show()
```
